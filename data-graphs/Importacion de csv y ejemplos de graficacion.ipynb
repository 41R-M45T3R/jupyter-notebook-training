# Import de datos de archivo csv y graficacion.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

# Se importa el archivo csv con la libreria pandas y se crea el data framework.

```python
data = pd.read_csv('SI6.csv',index_col='Distancia (m)')
data.head(10)
```

# Como si fuera MatLab

    Se crea la figura, se da titulo, y demas.

```python
plt.figure(1,figsize=(16,8))
plt.title('Ejemplo de titulo con diccionario',fontdict={'family':'times new roman','weight':'bold','fontsize':25,'color':'orange'})
#
plt.plot(data.Chalcedony,'k',label='Chalcedony')
plt.xlabel('Distancia en metros',fontsize=18)
plt.ylabel('valor en el eje y',fontsize=18)
#
plt.xticks(np.arange(0,210,step=1))
#
plt.xlim(0,20)
#
plt.grid()
plt.legend(fontsize='xx-large')
plt.show()
```

# Con el dataframework de pandas es mas facil graficar, simplemente accediendo a la columna de datos con data."nombre de la columna"
    Se puede visualizar los nombres de las columnas con:

```python
data.columns
```

## Se puede acceder a las columnas por indice sin tener que escribir todo el nombre.

```python
print(data.columns[0])
print(data.columns[3])
```

## Graficando toda la tabla accediendo mediante los indices de la columna

```python
plt.figure(figsize=(16,8))
plt.title('Tratando de graficar todo',fontdict={'fontsize':20})
#
for i in data.columns:
    print(i)
    plt.plot(data[i],label=i,)
#
plt.xticks(np.arange(0,210,step=5))
plt.yticks(np.arange(-10,10,step=0.1))
plt.xlim((0,200))
plt.ylim((-0.6,0.6))
#
plt.grid()
plt.legend(fontsize='xx-large',loc='lower right')
#lower,upper,center,left right
plt.show()
#
#para guardar la figura
#plt.save_fig('name.png',dpi=300)
```
