# Generacion caotica 2x^2 - 1 con el api de pylab

```python
valores=[]
k=0.54321
# desde el 1 al 42
for i in range(1,43):
    xn=2*pow(k,2)-1
    k=xn;
    valores.append(xn)
#
print(valores)
```

```python
figure(figsize=(16,4))
plot(valores)
title('Comportamiento caotico de 2x^2 -1')
xlabel('muestras')
ylabel('amplitud')
show()
```

# Ahora si utilicemos la libreria de matplotlib directamente

## Graficacion sencilla

```python
x = np.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8]) # left botom width height in range 0 to 1

axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titulaso')
plt.show()
```

## Insertar un plot dentro de un plot

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0, 1,1])
axes2 = fig.add_axes([0.1,0.5,0.5,0.3]) #left , bottom , widt , height

#main
axes1.plot(x,y,'r')

#insert
axes2.plot(valores,'b')

plt.show()
```

## un subplot, tamao de figura y dpi

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(8,4),dpi=100)

for ax in axes:
    ax.plot(valores, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig.tight_layout() # acomoda bien rico todo para que no se encimen cosas.
```
